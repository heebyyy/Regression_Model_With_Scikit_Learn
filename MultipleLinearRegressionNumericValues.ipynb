```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
automobile_df = pd.read_csv('dataset/auto_mpg_processed.csv')
automobile_df.sample(5)
```

```python
x = automobile_df[['displacement', 'horsepower','weight']]
y = automobile_df['mpg']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
linear_model = LinearRegression(normalize = True).fit(x_train, y_train)
```

```python
print('Training Score: ', linear_model.score(x_train, y_train))
```

```python
features = x_train.columns
coef = pd.Series(linear_model.coef_, features).sort_values()
print(coef)
```

```python
y_pred = linear_model.predict(x_test)
```

```python
print('Testing Score: ', r2_score(y_test, y_pred))
```

```python
plt.figure(figsize = (10, 6))
plt.plot(y_pred, label = 'Predicted')
plt.plot(y_test.values, label = 'Actual')
plt.ylabel('mpg')
plt.legend()
```

```python
x = automobile_df[['displacement', 'horsepower','weight', 'acceleration', 'cylinders','age']]
y = automobile_df['mpg']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

linear_model = LinearRegression(normalize = True).fit(x_train, y_train)

print('Training Score: ', linear_model.score(x_train, y_train))

features = x_train.columns
coef = pd.Series(linear_model.coef_, features).sort_values()
print('_' * 40)
print()
print('coefficients')
print(coef)

y_pred = linear_model.predict(x_test)

print('_' * 40)
print()
print('Testing Score: ', r2_score(y_test, y_pred))
```
