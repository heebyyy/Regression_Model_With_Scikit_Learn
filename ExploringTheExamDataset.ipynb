```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
exams_df = pd.read_csv('dataset/exams.csv')
exams_df.sample(5)
```

```python
 exams_df.describe()
```

```python
from sklearn import preprocessing
```

```python
exams_df[['math score']] = preprocessing.scale(exams_df[['math score']].astype('float64'))
exams_df[['reading score']] = preprocessing.scale(exams_df[['reading score']].astype('float64'))
exams_df[['writing score']] = preprocessing.scale(exams_df[['writing score']].astype('float64'))
```

```python
exams_df.head()
```

```python
exams_df.describe()
```

```python
exams_df['parental level of education'].unique()
```

```python
parental_level_of_education = [
    "associate's degree",
    'some high school',
    "bachelor's degree",
    'high school',
    'some college',
    "master's degree"
    ]
```

```python
label_encoding = preprocessing.LabelEncoder()

label_encoding = label_encoding.fit(parental_level_of_education)
```

```python
exams_df['parental level of education'] = \
    label_encoding.transform(exams_df['parental level of education'].astype(str))
```

```python
exams_df.head()
```

```python
exams_df = pd.get_dummies(exams_df, columns = ['race/ethnicity','gender', 'lunch', 'test preparation course'],
                          drop_first = True)
exams_df.head()
```

```python
exams_df.to_csv('dataset/exams_processed.csv', index = False)
```
